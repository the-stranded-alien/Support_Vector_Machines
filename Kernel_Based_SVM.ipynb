## Kernel Based Classification in SVM's

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_circles
from sklearn import svm
```

```python
X, Y = make_circles(n_samples=500, noise=0.02)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
# Using RBF Kernel
svc_rbf = svm.SVC(kernel='rbf')
svc_rbf.fit(X, Y)
print(svc_rbf.score(X, Y))
```

```python
# Using Linear Kernel
svc_lin = svm.SVC(kernel='linear')
svc_lin.fit(X, Y)
print(svc_lin.score(X, Y))
```

```python
# Using Polynomial Kernel
svc_pol = svm.SVC(kernel='poly')
svc_pol.fit(X, Y)
print(svc_pol.score(X, Y))
```

```python
# Using Sigmoid Kernel
svc_sig = svm.SVC(kernel='sigmoid')
svc_sig.fit(X, Y)
print(svc_sig.score(X, Y))
```

### Custom Kernel

```python
def custom_kernel(x1, x2):
    return np.square(np.dot(x1, x2.T))

svc_custom = svm.SVC(kernel=custom_kernel)
svc_custom.fit(X, Y)
print(svc_custom.score(X, Y))
```
